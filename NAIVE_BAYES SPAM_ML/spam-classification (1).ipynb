```python
nltk.download("stopwords") #it contain few words which are not important wehich we will remove from ourr text
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
from nltk.stem.porter import PorterStemmer #to get the root words of simliar words
```

```python
ps=PorterStemmer()
```

We will do some preproceesiong
- Removing , . capital letters to lowe letters and etc
- then we will encode our lettters

```python
corpus = []
for i in range(len(df)):
    # Step 1: Remove non-alphabetic characters
    review = re.sub('[^a-zA-Z]', ' ', df["messages"][i])
    print(f"Step 1 - Non-alphabetic characters removed: {review}")
    
    # Step 2: Convert to lowercase
    review = review.lower()
    print(f"Step 2 - Converted to lowercase: {review}")
    
    # Step 3: Split into words
    review = review.split()
    print(f"Step 3 - Split into words: {review}")
    
    # Step 4: Remove stopwords and stem the words
    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]
    print(f"Step 4 - Stopwords removed and words stemmed: {review}")
    
    # Step 5: Join the words back into a single string
    review = ' '.join(review)
    print(f"Step 5 - Words joined: {review}")
    
    # Step 6: Add to the corpus
    corpus.append(review)

# Output the processed corpus
print("Final corpus:")
print(corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
y=pd.get_dummies(df['label'],dtype=int,drop_first=True)
```

#### now we will create model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X1,y,random_state=42,test_size=25)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
